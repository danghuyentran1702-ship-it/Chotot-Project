```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import ast
import re

df_img = df[['price', 'image', 'image_thumbnails', 'images']].copy()

# Lọc price hợp lệ
df_img = df_img.dropna(subset=['price'])
df_img = df_img[df_img['price'] > 0]

# ===== Hàm đếm số ảnh trong cell (robust) =====
def count_items(x):
    if pd.isna(x):
        return 0

    # Nếu đã là list
    if isinstance(x, list):
        return len(x)

    # Nếu là dict
    if isinstance(x, dict):
        return len(x)

    # Nếu là string
    if isinstance(x, str):
        s = x.strip()
        if s == "" or s.lower() in ["nan", "none", "null"]:
            return 0

        # Nếu là JSON string list
        try:
            parsed = ast.literal_eval(s)
            if isinstance(parsed, list):
                return len(parsed)
            if isinstance(parsed, dict):
                return len(parsed)
        except:
            pass

        # Nếu là chuỗi URL phân cách bằng dấu phẩy
        if "," in s:
            return len([i for i in s.split(",") if i.strip() != ""])

        # Nếu là 1 URL
        if "http" in s:
            return 1

        # fallback: đếm số link trong text
        links = re.findall(r'http[s]?://\S+', s)
        return len(links)

    return 0

# ===== Tạo số lượng ảnh =====
df_img['num_image'] = df_img['image'].apply(count_items)
df_img['num_thumbnails'] = df_img['image_thumbnails'].apply(count_items)
df_img['num_images'] = df_img['images'].apply(count_items)

# ===== Tạo biến có/không =====
df_img['has_image'] = (df_img['num_image'] > 0).astype(int)
df_img['has_thumbnail'] = (df_img['num_thumbnails'] > 0).astype(int)
df_img['has_images'] = (df_img['num_images'] > 0).astype(int)

# log_price
df_img['log_price'] = np.log10(df_img['price'])

df_img[['num_image','num_thumbnails','num_images']].describe()
```

Một phần price_string không có giá rõ ràng → cần lọc hoặc xử lý riêng.

```python
#3.2 Parse giá từ price_string (tỷ / triệu)

import re

def parse_price_string_to_million(s):
    if pd.isna(s):
        return np.nan

    s = str(s).strip().lower()
    if s in ['nan', 'none', '']:
        return np.nan

    # loại cases không có giá
    if any(k in s for k in ['thoả thuận', 'liên hệ', 'đang cập nhật']):
        return np.nan

    # đổi dấu phẩy thành dấu chấm
    s = s.replace(',', '.')

    # tìm số (có thể có dấu chấm)
    match = re.search(r'(\d+(\.\d+)?)', s)
    if not match:
        return np.nan

    value = float(match.group(1))

    # nếu có "tỷ"
    if 'tỷ' in s or 'ty' in s:
        return value * 1000  # 1 tỷ = 1000 triệu

    # nếu có "triệu"
    if 'triệu' in s or 'trieu' in s:
        return value

    # nếu có "nghìn" -> triệu
    if 'nghìn' in s or 'nghin' in s:
        return value / 1000

    return np.nan
```

```python
#3.3 Tạo cột parsed_price_million từ price_string
df_p['parsed_price_million'] = df_p['price_string'].apply(parse_price_string_to_million)

df_p['parsed_price_million'].describe()
```

```python
def count_outliers(series):
    Q1 = series.quantile(0.25)
    Q3 = series.quantile(0.75)
    IQR = Q3 - Q1
    return ((series < Q1 - 1.5*IQR) | (series > Q3 + 1.5*IQR)).sum()

outlier_stats = (
    df_apartment_type_analysis.groupby('apartment_type_name')['log_price']
      .apply(count_outliers)
      .reset_index(name='outlier_count')
)

outlier_stats
```

```python
import pandas as pd
keywords_dict = {
    'is_chinh_chu': ['chính chủ', 'f0', 'chủ gửi', 'không qua trung gian'],
    'is_gia_re': ['giá rẻ', 'giá ngộp', 'cắt lỗ', 'thanh lý', 'thu hồi vốn'],
    'is_no_hau': ['nở hậu', 'phong thủy tốt'],
    'is_hem_xe_hoi': ['hẻm xe hơi', 'hẻm ôtô', 'hxh', 'xe hơi đỗ cửa']
}
def extract_keywords(text, keywords):
    if pd.isna(text):
        return 0
    text = text.lower()
    for word in keywords:
        if word in text:
            return 1
    return 0
for col_name, words in keywords_dict.items():
    df[col_name] = df['subject'].apply(lambda x: extract_keywords(x, words))
print("Thống kê từ khóa trích xuất được:")
for col in keywords_dict.keys():
    print(f"- {col}: {df[col].sum()} tin")

display(df[['subject'] + list(keywords_dict.keys())].head(10))
```

```python
import pandas as pd
df['street_number'] = df['street_number'].fillna('').astype(str)
street_name_map = df.groupby('unique_street_id')['street_name'].agg(lambda x: x.value_counts().index[0]).to_dict()
street_analysis = df.groupby('unique_street_id').agg({
    'sold_ads': 'sum',
    'price': 'median',
    'subject': 'count'
}).reset_index()
street_analysis['street_display_name'] = street_analysis['unique_street_id'].map(street_name_map)

print("Đã loại bỏ số nhà và tổng hợp dữ liệu theo Tuyến đường thành công!")
display(street_analysis.head())
```

```python
import matplotlib.pyplot as plt
street_sorted = street_analysis.sort_values(by='sold_ads', ascending=False)
street_sorted['cumulative_percent'] = 100 * street_sorted['sold_ads'].cumsum() / street_sorted['sold_ads'].sum()

fig, ax1 = plt.subplots(figsize=(12, 6))

ax1.bar(street_sorted['street_display_name'].head(15), street_sorted['sold_ads'].head(15), color='C0')
ax1.set_ylabel('Số lượng tin đã bán', color='C0')
ax1.tick_params(axis='x', rotation=45)
ax2 = ax1.twinx()
ax2.plot(street_sorted['street_display_name'].head(15), street_sorted['cumulative_percent'].head(15), color='C1', marker="D", ms=7)
ax2.set_ylabel('% Tích lũy', color='C1')
ax2.set_ylim(0, 105)

plt.title('Biểu đồ Pareto: Các tuyến đường đóng góp doanh số nhiều nhất', fontsize=15)
plt.show()
```
