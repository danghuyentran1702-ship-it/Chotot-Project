```python
RAW_DATA_PATH = "/content/drive/MyDrive/Chotot Project/RawData"
TRANSFORM_DATA_PATH = "/content/drive/MyDrive/Chotot Project/TransformData"
```

```python
import os
import json
import pandas as pd
```

# 1. Gom dữ liệu mua bán "s" cho tất cả các loại bất động sản

```python
subset_list = [
    "chotot_st_s_cg_1010/chotot_st_s_cg_1010_20251224_010402.json",
    "chotot_st_s_cg_1020/chotot_st_s_cg_1020_20251224_011615.json",
    "chotot_st_s_cg_1030/chotot_st_s_cg_1030_20251224_011638.json",
    "chotot_st_s_cg_1040/chotot_st_s_cg_1040_20251224_011907.json",
]
```

```python
all_data = []
for subset_path in subset_list:
  path = os.path.join(RAW_DATA_PATH, subset_path)
  with open(path) as f:
    data = json.load(f)
    all_data.extend(data)
```

```python
s_df = pd.DataFrame(all_data)
```

```python
print(len(s_df))
s_df.head()
```

```python
for column in s_df.columns:
  print(column)
```

```python
remove_fields = [
    "ad_features",
    "ad_id",
    "ad_labels",
    "avatar",
    "business_days",
    "date",
    "inspection_images",
    "label_campaigns",
    "params",
    "pty_characteristics",
    "specific_service_offered",
    "is_zalo_show",
    "seller_info",
    "is_sticky",
    "shop",
    "shop_alias",
    "projectimages",
    "phone_hidden",
]
```

```python
s_df = s_df.drop(columns=remove_fields)
```

```python
s_df.head()
```

```python
s_df.to_csv(
    os.path.join(TRANSFORM_DATA_PATH, "s_listing.csv"),
    index=False
)
```

```python
for col, dtype in s_df.dtypes.items():
    print(f"{col}: {dtype}")
```
